## Treinamento do Modelo

```python
caracteristicas = dados.iloc[:, 1:4].values
previsor = dados.iloc[:, 4:5].values
```

```python
# Separando o dataset: 80% para treino e 20% para teste
X_train, X_test, y_train, y_test = train_test_split(caracteristicas, previsor, test_size=0.20)
```

```python
# Instanciando modelo
funcao_logistica = LogisticRegression()
```

```python
# Treinando o modelo
funcao_logistica.fit(X_train, y_train)
```

## Avaliação do Modelo

```python
previsoes = funcao_logistica.predict(X_test)
previsoes
```

#### Matriz de Confusão

```python
confusion_matrix(y_test, previsoes)
```

#### Relatório de Classificação

```python
print(classification_report(y_test, previsoes))
# recall = acertividade
```

## Implementação do Modelo

```python
# Função para previsão
def previsao():
    salario = int(input('Salário em reais: '))
    tipo_renda = int(input('Tipo de renda: '))
    possui_imovel = int(input('Possui imóvel: '))
    parametro = [[salario, tipo_renda, possui_imovel]]
    previsao = funcao_logistica.predict(parametro)
    probabilidade = funcao_logistica.predict_proba(parametro)

    if previsao == 0:
        print()
        print('Não irá comprar')
        print('Probabilidade: {}'.format(probabilidade))
    else:
        print()
        print('Vai comprar')
        print('Probabilidade: {}'.format(probabilidade))
```
